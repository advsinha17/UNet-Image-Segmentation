```python
import os
import matplotlib.pyplot as plt
import random
```

```python
results_dir = os.path.join(os.getcwd(), "predictions")
imgs_dir = os.path.join(os.getcwd(), "data/val")
```

```python
num_imgs = 5
result_imgs = os.listdir(results_dir)
random.shuffle(result_imgs)
random_imgs = result_imgs[:num_imgs]
_, axs = plt.subplots(num_imgs, 2, figsize=(10, 15))
for i in range(num_imgs):
    true_img_dir = os.path.join(imgs_dir, result_imgs[i])
    pred_img_dir = os.path.join(results_dir, result_imgs[i])
    true_img = plt.imread(true_img_dir)
    pred_img = plt.imread(pred_img_dir)
    axs[i, 0].imshow(true_img)
    axs[i, 0].set_title("True")
    axs[i, 0].get_xaxis().set_visible(False)
    axs[i, 0].get_yaxis().set_visible(False)
    axs[i, 1].imshow(pred_img)
    axs[i, 1].set_title("Predicted")
    axs[i, 1].get_xaxis().set_visible(False)
    axs[i, 1].get_yaxis().set_visible(False)  
```
